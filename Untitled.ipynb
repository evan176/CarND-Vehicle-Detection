```python
import os

import numpy as np
from skimage.color import rgb2gray, rgb2hsv, rgb2ycbcr
from skimage.feature import hog
from skimage.exposure import histogram
from skimage.io import imread
from skimage.filters import sobel
from skimage.transform import resize
from sklearn.preprocessing import normalize
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from scipy.ndimage.measurements import label as label_func
import  matplotlib.pyplot as plt
import cv2
import skvideo.io
from tqdm import tqdm

%matplotlib inline 
```

```python
nonvehicle_path = "non-vehicles"
vehicle_path = "vehicles"
negative_imgs = []
positive_imgs = []

for root, dirs, files in os.walk(nonvehicle_path):
    for name in files:
        if name[-3:].lower() in ["jpg", "png"]:
            negative_imgs.append(os.path.join(root, name))

for root, dirs, files in os.walk(vehicle_path):
    for name in files:
        if name[-3:].lower() in ["jpg", "png"]:
            positive_imgs.append(os.path.join(root, name))
```

```python
def scale(img, output_shape=(64, 64)):
    return resize(img, output_shape)


def bin_spatial(img, size=(32, 32)):
    color1 = resize(img[:, :, 0], size).ravel()
    color2 = resize(img[:, :, 1], size).ravel()
    color3 = resize(img[:, :, 2], size).ravel()
    return np.hstack((color1, color2, color3))


def cal_sobel(img):
    return sobel(resize(rgb2gray(img), (32, 32))).ravel()


def cal_channel_hist(img, nbins=32):
    r_hist, r_center = histogram(img[:, :, :1], nbins=nbins)
    g_hist, g_center = histogram(img[:, :, 1:2], nbins=nbins)
    b_hist, b_center = histogram(img[:, :, 2:3], nbins=nbins)
    return np.hstack([r_hist, g_hist, b_hist])


def extract_feature(img, scale_shape=(64, 64), nbins=32, pix_per_cell=8,
                    cell_per_block=2, orient=9):
    img_bins = bin_spatial(img)
    
    img = scale(img, scale_shape)
    hog_features, hog_image = hog(
        rgb2gray(img), orientations=orient,
        pixels_per_cell=(pix_per_cell, pix_per_cell), 
        cells_per_block=(cell_per_block, cell_per_block),
        transform_sqrt=True,
        visualise=True, feature_vector=True,
        block_norm="L2-Hys"
    )
    rgb_hists = cal_channel_hist(img, nbins=nbins)
    hsv_hists = cal_channel_hist(rgb2hsv(img), nbins=nbins)
    ycbcr_hists = cal_channel_hist(rgb2ycbcr(img), nbins=nbins)
    features = np.hstack([hog_features, rgb_hists, hsv_hists, ycbcr_hists])
    
    features = np.hstack([[img_bins], normalize([features])])
    return features[0]
```

```python
negative_data = []
positive_data = []

for img_path in tqdm(negative_imgs):
    img = imread(img_path)
    negative_data.append(extract_feature(img))
    
for img_path in tqdm(positive_imgs):
    img = imread(img_path)
    positive_data.append(extract_feature(img))
    
    
negative_data = np.array(negative_data)
positive_data = np.array(positive_data)
```

```python
data = np.vstack([negative_data, positive_data])
label = [0] * len(negative_data) + [1] * len(positive_data)
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(data)
```

```python
pca = PCA(n_components=300)
data = pca.fit_transform(data)
print(sum(pca.explained_variance_ratio_))
```

```python
C = 1
label = np.array(label)
train_accs, valid_accs = [], []
kf = KFold(n_splits=10)
for train_index, valid_index in kf.split(data):
    clf = LogisticRegression(C=C, random_state=0, solver='lbfgs')
    # clf = SVC(kernel="linear", C=C)
    clf.fit(data[train_index], label[train_index])
    train_acc = accuracy_score(label[train_index], clf.predict(data[train_index]))
    valid_acc = accuracy_score(label[valid_index], clf.predict(data[valid_index]))
    train_accs.append(train_acc)
    valid_accs.append(valid_acc)
    print("train: {}, valid: {}".format(train_acc, valid_acc))
print("Mean - train: {}, valid: {}".format(sum(train_accs) / 10.0, sum(valid_accs) / 10.0))
```

```python
clf = LogisticRegression(C=C)
# clf = SVC(kernel="linear", C=C)
clf.fit(data, label)
```

```python
def slide_window(img, ystart, ystop, xstart, xstop, box_size, overlap_size):
    window_inds = []
    for y in range(ystart, ystop - overlap_size, overlap_size):
        for x in range(xstart, xstop - overlap_size, overlap_size):
            window_inds.append((y, x))
    return window_inds


def update_heatmap(img, window_inds, box_size, heatmap):
    for y, x in window_inds:
        features = extract_feature(img[y:y+box_size, x:x+box_size])
        features = scaler.transform([features])
        features = pca.transform(features)
        prob = clf.predict_proba(features)
        # heatmap[y:y+box_size, x:x+box_size] += prob[0][1]
        if prob[0][1] > 0.5:
            heatmap[y:y+box_size, x:x+box_size] += 1
    return heatmap


def find_cars(img):
    height, width, channels = img.shape
    heatmap = np.zeros((height, width))
    
    window_inds = slide_window(img, 400, 500, 0, width, 64, 32)
    heatmap = update_heatmap(img, window_inds, 64, heatmap)
    window_inds = slide_window(img, 400, 600, 0, width, 96, 48)
    heatmap = update_heatmap(img, window_inds, 128, heatmap)
    window_inds = slide_window(img, 400, 650, 0, width, 128, 64)
    heatmap = update_heatmap(img, window_inds, 128, heatmap)
    return heatmap


def label_cars(heatmap, threshold):
    new_map = np.copy(heatmap)
    new_map[heatmap <= threshold] = 0
    labels = label_func(new_map)
    return labels


def draw_labeled_bboxes(img, labels):
    # Iterate through all detected cars
    img = np.copy(img)
    for car_number in range(1, labels[1]+1):
        # Find pixels with each car_number label value
        nonzero = (labels[0] == car_number).nonzero()
        # Identify x and y values of those pixels
        nonzeroy = np.array(nonzero[0])
        nonzerox = np.array(nonzero[1])
        # Define a bounding box based on min/max x and y
        bbox = ((np.min(nonzerox), np.min(nonzeroy)), (np.max(nonzerox), np.max(nonzeroy)))
        # Draw the box on the image
        cv2.rectangle(img, bbox[0], bbox[1], (0, 0, 255), 6)
    # Return the image
    return img
```

```python
# Read in the last image above
img = imread('test_images/test1.jpg')
# Draw bounding boxes on a copy of the image
heatmap = find_cars(img)
plt.imshow(heatmap)
plt.show()
```

```python
labels = label_cars(heatmap, 2)
draw_img = draw_labeled_bboxes(img, labels)
# Display the image
plt.imshow(draw_img)
plt.show()
```

```python
heatmaps = []

cap = cv2.VideoCapture("project_video.mp4")
frame_len = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
out = skvideo.io.FFmpegWriter("project_out.mp4")
for i in tqdm(range(frame_len)):
    if not cap.isOpened():
        break
    ret, frame = cap.read()
    if ret:
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        heatmap = find_cars(img)
        heatmaps.append(heatmap)
        if len(heatmaps) > 10:
            heatmaps.pop(0)
        labels = label_cars(np.mean(heatmaps, axis=0), 2)
        draw_img = draw_labeled_bboxes(img, labels)
        out.writeFrame(draw_img)          
out.close()
cap.release()
```
